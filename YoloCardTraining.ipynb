```python
from ultralytics import YOLO
from pathlib import Path
import yaml
```

```python
# Load and display the dataset config
DATA_YAML = Path("CardDataset/data.yaml")

with open(DATA_YAML) as f:
    data_config = yaml.safe_load(f)

print("Dataset Configuration:")
print(f"  Number of classes: {data_config['nc']}")
print(f"  Class names: {data_config['names']}")

# Count images in each split
for split in ["train", "valid", "test"]:
    img_dir = Path(f"CardDataset/{split}/images")
    if img_dir.exists():
        count = len(list(img_dir.iterdir()))
        print(f"  {split}: {count} images")
```

```python
model = YOLO("yolo26m.pt")
```

```python
# train on the playing cards dataset (52 classes)
results = model.train(
    data=str(DATA_YAML.resolve()),
    epochs=6,
    imgsz=448,
    batch=8,           # safe batch size
    patience=10,        # early stopping patience
    device=0,           # use GPU
    workers=4,          
    project=str(Path("runs/cards").resolve()),
    name="yolo26m_cards",
    verbose=True,
)
```

```python
model.save("PlayingCardsM.pt")
```
